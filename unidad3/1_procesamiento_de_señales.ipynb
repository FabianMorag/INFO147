# Señales

¿Qué es una señal?

> Es una representación de la **variación** de los **atributos de un sistema**

Las variaciones se miden con respecto a una **variable independiente**

El valor de los atributos del sistema **depende** de las variables independientes

> La señal es una relación entre las variables independientes y dependientes

En matemáticas, este concepto se llama **función**:

$$
f(x) = y,
$$

donde $x$ es la variable independiente (entrada) e $y$ es la variable dependiente (salida)

# Ejemplos

La definición de señal es muy amplia, podemos concretar la idea con ejemplos

- Economía: Índice financiero en función de la hora 

- Meteorología: Presión y temperatura en función de la posición espacial

- Electrónica: Nivel lógico de una compuerta en función del tiempo

- Astronomía: Brillo de una estrella pulsante en función del instante de observación (Curva de luz)

- Biología: Voltaje en la superficie de craneo en función del tiempo (Electro-encefalograma)



# Introducción al procesamiento de señales

¿Qué es procesamiento de señales?

> Desarrollo y uso de técnicas para detectar, extraer, filtrar, transformar, transmitir y reproducir señales

¿Qué es el procesamiento digital de señales?

> Uso de algoritmos y modelos para procesar señales digitales con un computador

¿Qué es una señal digital?

> Es una señal que ha sido discretizada y convertida en **datos**
